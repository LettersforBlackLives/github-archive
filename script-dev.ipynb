```python
import pandas as pd
import re
import os
import subprocess
```

```python
data = pd.read_csv("2020LettersforBlackLives.tsv", sep = '\t')
```

```python
data.head()
```

```python
with_links = data.iloc[data['Published on Medium (Link)'].dropna().index]
```

```python
with_links
```

```python
def fmt_filename(s):
    try:
        s = re.sub('[\W_]+', '_',s)
    except:
        pass
    return s + '.md'
```

```python
with_links['fname'] = with_links['Translation document (link)'].apply(fmt_filename)
```
